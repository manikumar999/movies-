```python
# importing dependencies
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
import numpy as np
```

```python
columns=['Year','Title', 'Director', 'Cast', 'Music director', 'Sources']
df = pd.DataFrame(columns = columns)
df
```

```python
for i in range(1992 , 2000):
    html = requests.get('https://en.wikipedia.org/wiki/List_of_Telugu_films_of_'+str(i)).text
# lets create a beatifulsoup object with html.parser structure
    bsobj = bs(html , 'html.parser')
    table = bsobj.find('table' , {'class':'wikitable'}).tbody
    rows = table.find_all('tr')
   # columns = [x.text.replace('\n','') for x in rows[0].find_all('th')]
    #columns.append('Year')
    for j in range(1 , len(rows)):
        values = [x.text.replace('\n','') for x in rows[j].find_all('td')]
        values.insert(0,i)
        df=df.append(pd.Series(values , index=columns) , ignore_index=True)
    
    
```

```python
df.drop(['Sources','Music director'],axis=1, inplace=True)
```

```python
df1 = pd.DataFrame(columns = columns1)
for i in range(2000,2010):
    if i != 2004:
        html = requests.get('https://en.wikipedia.org/wiki/List_of_Telugu_films_of_'+str(i)).text
        bsobj =bs(html,'html.parser')
        table =bsobj.find_all('table',class_='wikitable', style_=False)

        for a in table:
            rows = a.tbody.find_all('tr')
            for j in range(1 , len(rows)):
                values=[]
                for x in rows[j].find_all('td'):
                    try:
                        float(x.text.replace('\n',''))
                    except ValueError:
                        values.append(x.text.replace('\n',''))
            
           
                for k in ['JAN',"FEB","MAR",'APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']:
                    if k in values:
                        count=k
                if count not in values:   
                    values.insert(0,count)
    #values = [x.text.replace('\n','') for x in rows[j].find_all('td') 
    #values.append(2000)
                values.insert(0,i)
                if len(values)==7:
                    df1=df1.append(pd.Series(values , index=columns1) , ignore_index=True)
                
```
